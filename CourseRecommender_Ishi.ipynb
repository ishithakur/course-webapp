```python
data = data[['Course Name','Difficulty Level','Course Description','Skills','Course Rating']]
data.head(5)
```

```python
# Removing spaces between the words (Lambda funtions can be used as well)

data['Course Name'] = data['Course Name'].str.replace(' ',',')
data['Course Name'] = data['Course Name'].str.replace(',,',',')
data['Course Name'] = data['Course Name'].str.replace(':','')
data['Course Description'] = data['Course Description'].str.replace(' ',',')
data['Course Description'] = data['Course Description'].str.replace(',,',',')
data['Course Description'] = data['Course Description'].str.replace('_','')
data['Course Description'] = data['Course Description'].str.replace(':','')
data['Course Description'] = data['Course Description'].str.replace('(','')
data['Course Description'] = data['Course Description'].str.replace(')','')

#removing paranthesis from skills columns
data['Skills'] = data['Skills'].str.replace('(','')
data['Skills'] = data['Skills'].str.replace(')','')
```

```python
data['tags'] = data['Course Name'] + data['Difficulty Level'] + data['Course Description'] + data['Skills']
data.head(5)
```

```python
data['tags'].iloc[1]
```

```python
new_data = data[['Course Name','tags']]
new_data.head(5)
```

```python
new_data['tags'] = data['tags'].str.replace(',',' ')
```

```python
new_data['Course Name'] = data['Course Name'].str.replace(',',' ')
```

```python
new_data.rename(columns = {'Course Name':'course_name'}, inplace = True)
```

```python
new_data['tags'] = new_data['tags'].apply(lambda x:x.lower()) #lower casing the tags column
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_data['tags']).toarray()
```

```python
import nltk #for stemming process
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
#defining the stemming function
def stem(text):
    y=[]

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
#new_data['tags'] = new_data['tags'].apply(stem) #applying stemming on the tags column
stem
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer

# Assuming you have a DataFrame 'df' with 'text' column containing text data and 'label' column for labels
X = new_data['tags']  # Text data
y = new_data['course_name']  # Labels

# Convert text data to numerical features using CountVectorizer (or any other vectorization method)
vectorizer = CountVectorizer()
X_vec = vectorizer.fit_transform(X)

# Split the data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_vec, y, test_size=0.2, random_state=42)

# Now you have X_train and y_train for training your NLP model, and X_test and y_test for testing/evaluation.
```

```python
def recommend(course):
    course_index = new_data[new_data['course_name'] == course].index[0]
    distances = similarity[course_index]
    course_list = sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:7]

    for i in course_list:
        print(new_data.iloc[i[0]].course_name)
```

```python
import random
recommend(random.choice(new_data['course_name']))
```
